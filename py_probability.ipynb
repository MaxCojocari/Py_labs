## **Densities** 

At a mathematical conference, 10 participants are randomly seated around a circular table for meals. Using simulation, estimate the probability that no two people sit next to each other at both lunch and dinner. Can you make an intelligent conjecture for the case of n participants when n is large?

```python
import matplotlib.pyplot as plt
import random
```

```python
def gen_neighbours(A, neighbours = {}):
    N = len(A)
    neighbours.update({A[0]:{A[-1], A[1]}})
    for i in range(1, N-1):
        neighbours.update({A[i]: {A[i-1], A[i+1]}})
    neighbours.update({A[N-1]:{A[N-2], A[0]}})
    return neighbours
```

```python
def find_probability(N, nr_exp):
    A = [x for x in range(1, N+1)]
    B = list(A)
    k = 0
    for i in range(nr_exp):
        random.shuffle(A)
        random.shuffle(B)
        neighbour_dict_reference = dict(gen_neighbours(A))
        neighbour_dict = gen_neighbours(B)
        w = 0
        for i in range(1, N+1):
            if neighbour_dict[i].isdisjoint(neighbour_dict_reference[i]):
                w += 1
        if w == N:
            k += 1
    return k/nr_exp
```

```python
N = int(input())
print(find_probability(N, 1000000))
```

```python
number_of_participants = [x for x in range(2, 101)]
probability = [find_probability(n, 10000) for n in number_of_participants]
```

```python
plt.figure(dpi=150)
plt.plot(number_of_participants, probability, color='red')
plt.xlabel('Nr participants')
plt.ylabel('Probability')
plt.title('N = 10000')
plt.grid(True)
plt.show()
```
